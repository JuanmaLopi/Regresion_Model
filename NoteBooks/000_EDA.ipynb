```python
import pandas as pd
```

```python
# Cargar los archivos CSV
data_2015 = pd.read_csv('../data/2015.csv')
data_2016 = pd.read_csv('../data/2016.csv')
data_2017 = pd.read_csv('../data/2017.csv')
data_2018 = pd.read_csv('../data/2018.csv')
data_2019 = pd.read_csv('../data/2019.csv')
```

```python
print("Columnas en 2015.csv:", data_2015.columns)
print("Columnas en 2016.csv:", data_2016.columns)
print("Columnas en 2017.csv:", data_2017.columns)
print("Columnas en 2018.csv:", data_2018.columns)
print("Columnas en 2019.csv:", data_2019.columns)
```

```python
data_2015 = data_2015.drop(columns={
    'Country',
    'Region',
    'Happiness Rank',
    'Standard Error',
    'Dystopia Residual'
})

data_2015 = data_2015.rename(columns={
    'Family': 'Social support'
})
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la matriz de correlacin
correlation_matrix = data_2015.corr()

# Crear un mapa de calor de la matriz de correlacin
plt.figure(figsize=(10, 8))  # Ajusta el tamao de la figura
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

# Personalizar el grfico
plt.title('Mapa de calor de correlacin')
plt.show()
```

```python
data_2016 = data_2016.drop(columns={
    'Country',
    'Region',
    'Happiness Rank',
    'Dystopia Residual',
    'Lower Confidence Interval',
    'Upper Confidence Interval'
})

data_2016 = data_2016.rename(columns={
    'Family': 'Social support'
})
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la matriz de correlacin
correlation_matrix = data_2016.corr()

# Crear un mapa de calor de la matriz de correlacin
plt.figure(figsize=(10, 8))  # Ajusta el tamao de la figura
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

# Personalizar el grfico
plt.title('Mapa de calor de correlacin')
plt.show()
```

```python
data_2017.columns = data_2017.columns.str.replace('"', '', regex=False).str.replace("'", "", regex=False)


data_2017 = data_2017.drop(columns={
    'Country',
    'Happiness.Rank',
    'Whisker.high',
    'Whisker.low',
    'Dystopia.Residual'

})

data_2017 = data_2017.rename(columns={
    'Happiness.Score': 'Happiness Score',
    'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)',
    'Family': 'Social support',
    'Health..Life.Expectancy.': 'Health (Life Expectancy)',
    'Trust..Government.Corruption.': 'Trust (Government Corruption)'
})

data_2017['Happiness Score'] = data_2017['Happiness Score'].round(3)

# Verificar el cambio en los nombres de las columnas
print(data_2017.columns)
```

```python
columnas = list(data_2017.columns)
columnas[5], columnas[6] = columnas[6], columnas[5] 
data_2017 = data_2017[columnas]
print(data_2017.columns)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la matriz de correlacin
correlation_matrix = data_2017.corr()

# Crear un mapa de calor de la matriz de correlacin
plt.figure(figsize=(10, 8))  # Ajusta el tamao de la figura
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

# Personalizar el grfico
plt.title('Mapa de calor de correlacin')
plt.show()
```
