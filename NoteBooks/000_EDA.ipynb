```python
import pandas as pd
```

```python
# Cargar los archivos CSV
data_2015 = pd.read_csv('../data/2015.csv')
data_2016 = pd.read_csv('../data/2016.csv')
data_2017 = pd.read_csv('../data/2017.csv')
data_2018 = pd.read_csv('../data/2018.csv')
data_2019 = pd.read_csv('../data/2019.csv')
```

```python
print("Columnas en 2015.csv:", data_2015.columns)
print("Columnas en 2016.csv:", data_2016.columns)
print("Columnas en 2017.csv:", data_2017.columns)
print("Columnas en 2018.csv:", data_2018.columns)
print("Columnas en 2019.csv:", data_2019.columns)
```

```python
data_2015 = data_2015.drop(columns={
    'Region',
    'Happiness Rank',
    'Standard Error',
    'Dystopia Residual'
})

data_2015 = data_2015.rename(columns={
    'Family': 'Social support'
})

data_2015 = data_2015.round(4)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la matriz de correlación de las variables numericas
correlation_matrix = data_2015.drop(columns=['Country']).corr()

# Crear un mapa de calor de la matriz de correlación
plt.figure(figsize=(10, 8))  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

# Personalizar el gráfico
plt.title('Mapa de calor de correlación')
plt.show()
```

```python
data_2016 = data_2016.drop(columns={
    'Region',
    'Happiness Rank',
    'Dystopia Residual',
    'Lower Confidence Interval',
    'Upper Confidence Interval'
})

data_2016 = data_2016.rename(columns={
    'Family': 'Social support'
})

data_2016 = data_2016.round(4)
```
