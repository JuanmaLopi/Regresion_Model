### Library Import

In this step, we import the pandas library for data handling. pandas is a fundamental tool in Python for data manipulation and analysis, as it provides data structures like DataFrames that allow efficient management of large amounts of data.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the CSV Files

Next, we load the CSV files corresponding to the data for each year (2015-2019) into pandas DataFrame variables. This step allows us to read the data from the files and store them in structures that are easy to manipulate and analyze.


```python
# Load the CSV Files
data_2015 = pd.read_csv('../data/2015.csv')
data_2016 = pd.read_csv('../data/2016.csv')
data_2017 = pd.read_csv('../data/2017.csv')
data_2018 = pd.read_csv('../data/2018.csv')
data_2019 = pd.read_csv('../data/2019.csv')
```

### Inspecting the Columns of the CSV Files

To verify the structure of the loaded files, we inspect the columns present in each of the datasets. This helps ensure that the files have been loaded correctly and provides insight into the available variables for analysis.


```python
print("Columns in 2015.csv:", data_2015.columns)
print("Columns in 2016.csv:", data_2016.columns)
print("Columns in 2017.csv:", data_2017.columns)
print("Columns in 2018.csv:", data_2018.columns)
print("Columns in 2019.csv:", data_2019.columns)
```

### Data Cleaning and Transformation for 2015

In this step, we perform several cleaning operations on the 2015 dataset. First, we remove unnecessary columns for the analysis. Then, we rename a column to improve clarity. Additionally, we round the numerical values to 4 decimal places for easier interpretation. Here's how you can implement these steps:


```python
data_2015 = data_2015.drop(columns={
    'Region',
    'Happiness Rank',
    'Standard Error',
    'Dystopia Residual'
})

data_2015 = data_2015.rename(columns={
    'Family': 'Social support'
})

data_2015 = data_2015.round(4)
```

### Correlation Analysis between Variables

In this section, we calculate the correlation matrix of the numerical variables in the 2015 dataset, excluding the Country column, which does not have a direct relationship with the numerical variables. Then, we visualize this matrix using a heatmap to identify the strongest relationships between variables.

The heatmap will help us see which variables are highly correlated, which is useful for understanding interdependencies between features and identifying potential redundancies.


```python
# Calculate the correlation matrix for the numerical variables
correlation_matrix = data_2015.drop(columns=['Country']).corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

# Customize the plot
plt.title('Correlation Heatmap')
plt.show()

```

### Data Preprocessing for the Year 2016

In this step, we apply some additional transformations to the 2016 dataset. Specifically, we remove several columns that are not necessary for our analysis or model, and we rename the `Family` column to maintain consistency with the 2015 dataset.

The columns removed are:

- `Region`: Information about the region, which is not relevant for the analysis.
- `Happiness Rank`: The happiness ranking, which does not provide additional information for the numerical analysis.
- `Dystopia Residual`: A residual value from dystopia, which is not used in the analysis.
- `Lower Confidence Interval and Upper Confidence Interval`: Confidence intervals that are not needed for the model.

Finally, we round the numerical values to 4 decimal places to maintain precision and avoid issues with small variations during the analysis.


```python
data_2016 = data_2016.drop(columns={
    'Region',
    'Happiness Rank',
    'Dystopia Residual',
    'Lower Confidence Interval',
    'Upper Confidence Interval'
})

data_2016 = data_2016.rename(columns={
    'Family': 'Social support'
})

data_2016 = data_2016.round(4)
```

### Correlation Analysis for the Year 2016

In this step, we calculate the correlation matrix between the numerical variables in the 2016 dataset. This helps us identify how the different variables are related to each other, which can be useful for understanding the interdependencies within the happiness model.

The `Country` column is removed to ensure that only numerical variables are analyzed. Then, a heatmap is created to visualize these correlations. A heatmap is a visual tool that uses colors to represent the correlation values between variables.

The code to calculate and visualize the correlation matrix is as follows:


```python
# Calculate the correlation matrix for the numerical variables
correlation_matrix = data_2016.drop(columns=['Country']).corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

# Customize the plot
plt.title('Correlation Heatmap')
plt.show()
```

### Data Cleaning and Preprocessing for the Year 2017

In this block of code, several data cleaning operations are performed on the 2017 dataset. These include the removal of unnecessary quotes from column names, elimination of irrelevant columns, and renaming of certain columns to improve their clarity.

First, double and single quotes are removed from column names using the `str.replace()` method. Then, columns that are not necessary for our analysis are dropped, such as `Happiness.Rank`, `Whisker.high`, `Whisker.low`, and `Dystopia.Residual`.

After that, some columns are renamed to make their names more descriptive and consistent with the other years' datasets. For example, `Happiness.Score` is renamed to `Happiness Score`, and `Economy..GDP.per.Capita`. is renamed to `Economy (GDP per Capita)`.

Finally, the data is rounded to 4 decimal places to ensure that the precision is appropriate for subsequent analysis and modeling.


```python
data_2017.columns = data_2017.columns.str.replace('"', '', regex=False).str.replace("'", "", regex=False)


data_2017 = data_2017.drop(columns={
    'Happiness.Rank',
    'Whisker.high',
    'Whisker.low',
    'Dystopia.Residual'

})

data_2017 = data_2017.rename(columns={
    'Happiness.Score': 'Happiness Score',
    'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)',
    'Family': 'Social support',
    'Health..Life.Expectancy.': 'Health (Life Expectancy)',
    'Trust..Government.Corruption.': 'Trust (Government Corruption)'
})

data_2017 = data_2017.round(4)

```

### Rearranging Columns in the 2017 Dataset

In this block of code, the columns of the 2017 dataset are reorganized. The goal is to swap the positions of two specific columns, in this case, the column in position 6 and the column in position 7. This operation is useful when the columns are not in the desired order for analysis.

The column reorganization is performed by converting the column names into a list, which is mutable. Then, the elements in positions 6 and 7 of the list (corresponding to the columns) are swapped. Finally, the modified column order is reassigned to the DataFrame.


```python
columnas = list(data_2017.columns)
columnas[6], columnas[7] = columnas[7], columnas[6] 
data_2017 = data_2017[columnas]
print(data_2017.columns)
```

### Correlation Analysis in the 2017 Dataset

In this block of code, the **correlation matrix** of the numerical variables in the 2017 dataset is calculated and visualized using a heatmap. The correlation matrix helps us understand how the different numerical variables relate to each other. Correlation values range from -1 to 1, where a value close to 1 indicates a strong positive relationship, and a value close to -1 indicates a strong negative relationship.

#### Steps Performed in the Code:

1. **Calculation of the Correlation Matrix**:
   - The "Country" column is excluded since it is not numerical, and it doesn't make sense to calculate correlations with it.
   - The `.corr()` method is then used to calculate the correlation matrix, which provides a matrix of correlation coefficients for the numerical columns in the DataFrame.

2. **Visualization of the Correlation Matrix with a Heatmap**:
   - The `seaborn` library is used to create a heatmap, which is a visual representation of the correlation matrix.
   - The `annot=True` option is used to display the numerical values in the cells of the heatmap.
   - The `coolwarm` color palette is chosen to represent the correlation values, where warm colors indicate strong correlations, and cool colors indicate weak correlations.
   - The `center=0` parameter ensures that the color scale is centered at 0, making it easier to interpret positive and negative correlations.

3. **Customization and Display**:
   - The figure size is adjusted using `figsize` to ensure that the heatmap is large enough to be readable.
   - Finally, the title of the plot is set, and the heatmap is displayed using `plt.show()`.


```python
# Calculate the correlation matrix for the numerical variables
correlation_matrix = data_2017.drop(columns=['Country']).corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

# Customize the plot
plt.title('Correlation Heatmap')
plt.show()
```

### Combining CSV Files

Once the data from the different years (2015-2019) has been cleaned and the columns have been renamed, the next step is to combine all the files into a single DataFrame for easier analysis and model construction. We use `pd.concat` to concatenate the DataFrames from the different years, and then remove rows with missing values using `dropna()`. Finally, we save the combined DataFrame into a new CSV file called `Model_Data.csv`.

Here is the code used to combine and save the files:

```python
# Combine the files
data_to_model = pd.concat([data_2015, data_2016, data_2017, data_2018, data_2019], ignore_index=True)
data_to_model = data_to_model.dropna()

# Save the combined file into a new CSV
data_to_model.to_csv('../data/Model_Data.csv', index=False)

print(f"Files successfully combined into 'Model_Data.csv' with {data_to_model.shape[0]} rows and {data_to_model.shape[1]} columns")
```
